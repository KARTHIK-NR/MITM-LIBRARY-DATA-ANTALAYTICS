```python

import pandas as pd
import matplotlib.pyplot as plt
data2020=pd.DataFrame()
data2021=pd.DataFrame()
data2022=pd.DataFrame()
data2020=pd.read_excel("C:/Users/dell/Downloads/MCA students transaction data/From August 2019 to July 2020.xlsx")
data2022=pd.read_csv("C:/Users/dell/OneDrive/Documents/2022_csv.csv")
data2021=pd.read_csv("D:/Downloards/2021.csv")
```

# hightest taken the books in each year

```python
# hightest taken the books in each year

x = data2022[data2022['Profile'].str.contains('Students')].groupby(["Gender"])["Gender"].count()
y = data2021[data2021['Profile'].str.contains('Students')].groupby(["Gender"])["Gender"].count()
z = data2020[data2020['Profile'].str.contains('Students')].groupby(["Gender"])["Gender"].count()

gender =['2019-20_', '_2020-21_','_2021-22_']

female=x[1]
male=x[2]
female1=y[1]
male1=y[2]
female2=z[1]
male2=z[2]

fig, ax = plt.subplots()
count=[male2+female2,male1+female1,male+female]

ax.bar(gender,count,width=0.5)
ax.set_xlabel('Count')
ax.set_ylabel('Transcation ')
ax.set_title('Number of Students  Transcation According to Year Wise')
fig.set_size_inches(14, 8)
for i in range(len(count)):
    plt.text(i, count[i], str(count[i]), ha='center', va='bottom')
plt.show()


```

# which branch taken the highest books in the the each year 

```python

#which branch taken the highest book in the the each year

branch2020=dict(data2020[data2020['Profile'].str.contains('Students')].groupby(["Branch"])["Branch"].count())
branch2021=dict(data2021[data2021['Profile'].str.contains('Students')].groupby(["Department"])["Department"].count())
branch2022=dict(data2022[data2022['Profile'].str.contains('Students')].groupby(["Branch"])["Branch"].count())


#we sort the element based the value of the each departments
res = {key: val for key, val in sorted(branch2020.items(), key = lambda ele: ele[1], reverse = True)}
res1 = {key: val for key, val in sorted(branch2021.items(), key = lambda ele: ele[1], reverse = True)}
res2 = {key: val for key, val in sorted(branch2022.items(), key = lambda ele: ele[1], reverse = True)}

#year of 2020
highest_branch_2020=list(res.keys())[0]
highest_value_2020=list(res.values())[0]
#year of 2021
highest_branch_2021=list(res1.keys())[0]
highest_value_2021=list(res1.values())[0]
#year of 2022
highest_branch_2022=list(res2.keys())[0]
highest_value_2022=list(res2.values())[0]
#ploting the dat using matplotlib
labels=[str(highest_branch_2020),str(highest_branch_2021),str(highest_branch_2022)]
counts=[highest_value_2020,highest_value_2021,highest_value_2022]


fig, ax = plt.subplots()
plt.bar(labels, counts)
plt.xlabel('branch')
plt.ylabel('number of student taken by each branch of students')
plt.title('departments')
fig.set_size_inches(16, 10)
for i in range(len(counts)):
    plt.text(i, counts[i], str(counts[i]), ha='center', va='bottom')
plt.show()
labels
```

```python
#how much fine collected by the each year

data2020["Fine"] = data2020["Fine"].apply(lambda x: int(x) if str(x).isdigit() else 0)
sum1 = data2020["Fine"].sum()

data2021["Fine"] = data2021["Fine"].apply(lambda x: int(x) if str(x).isdigit() else 0)
sum2 = data2021["Fine"].sum()

data2022["Fine"] = data2022["Fine"].apply(lambda x: int(x) if str(x).isdigit() else 0)
sum3 = data2022["Fine"].sum()

print(sum1,sum2,sum3)
```

# Year Wise  over all fines Students Have to Pay

```python
year = ['2020', '2021', '2022']
fig, ab = plt.subplots()
count = [sum1, sum2, sum3]
ab.bar(year, count, width=0.4)
ab.set_xlabel('Years')
ab.set_ylabel('Total Fines')
ab.set_title('Year Wise  over all fines Students Have to Pay')
fig.set_size_inches(10, 6)
for i in range(len(count)):
    plt.text(i, count[i], str(count[i]), ha='center', va='bottom')
plt.show()
```

# Trancations of Teaching and Non teaching Staff-2020

```python
import numpy as np
import seaborn as sns
total_number_of_faculty=dict(data2020[~ data2020['Profile'].str.contains('Students')].groupby("Profile")["Member Id"].nunique())
techer_datas = {key: val for key, val in sorted(total_number_of_faculty.items(), key = lambda ele: ele[1], reverse = True)}
x=list(techer_datas.keys())
yy=list(techer_datas.values())
s=[]
for i in range(len(x)):
    s.append(x[i].strip())
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=s,y=yy)
plt.xticks(rotation=45, ha='right')

plt.title("Trancations of Teaching and Non teaching Staff-2020")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
for i in range(len(yy)):
    plt.text(i, yy[i], str(yy[i]), ha='center', va='bottom')

plt.show()






```
